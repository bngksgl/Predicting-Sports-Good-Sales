```python
import pandas as pd
import numpy as np

df_items = pd.read_csv("items.csv",delimiter='|')
df_items[0:1]
```

### Data type

```python
# ajust data type
df_items['pid']=df_items['pid'].astype(str) # int-> str
#df_items['releaseDate']=pd.to_datetime(df_items["releaseDate"]) # object-> datetime
df_items.dtypes
```

### feature visualisation

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from pylab import *

sns.countplot(x="pid", data=df_items, palette="GnBu_d") 
sns.set(font_scale=2) #font size
sns.set_style("whitegrid")
#seaborn.set(rc={'axes.facecolor':'cornflowerblue', 'figure.facecolor':'cornflowerblue'})
rcParams['figure.figsize'] = 20,5
plt.show()
```

```python
sns.countplot(x="color", data=df_items, palette="GnBu_d") 
sns.set(font_scale=2) #font size
sns.set_style("whitegrid")
rcParams['figure.figsize'] = 20,5
plt.show()
```

```python
b=sns.countplot(x="brand", data=df_items, palette="GnBu_d") 
sns.set(font_scale=1)
sns.set_style("whitegrid")
rcParams['figure.figsize'] = 22,5
plt.show()
```

```python
sns.countplot(x="mainCategory", data=df_items, palette="GnBu_d") 
sns.set(font_scale=1.5)
sns.set_style("whitegrid")
rcParams['figure.figsize'] = 5,3
plt.show()
```

```python
sns.countplot(x="category", data=df_items, palette="GnBu_d") 
sns.set(font_scale=1.5)
sns.set_style("whitegrid")
rcParams['figure.figsize'] = 5,3
plt.show()
```
