### feature visualisation

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from pylab import *

sns.countplot(x="date", data=df_train, palette="GnBu_d") 
plt.xticks(rotation=90) #virtical x axis
sns.set(font_scale=1.5)
rcParams['figure.figsize'] = 30,8
plt.show()

# best sell: christmas 
```

```python
sns.countplot(x="units", data=df_train, palette="GnBu_d") 
sns.set(font_scale=1) #font size
rcParams['figure.figsize'] = 30,8
plt.show()
```

```python
sns.set(style="whitegrid", palette="GnBu_d")#color_codes=True)
sns.jointplot(x="units", y="pid", data=df_train, size=8, ratio=5)
plt.show()
```
